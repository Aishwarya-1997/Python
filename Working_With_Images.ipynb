<h1>CROPPING IMAGES : .crop(a,b,c,d)</h1> <h3>Takes A Tuple (a,b,c,d) - a and b are x and y co-ordinates respectively and c and d are width and height respectively</h3>

```python
mac.crop((0,0,100,100))
```

```python
mac.crop((0,0,200,150))
```

```python
mac.crop((5,0,150,150))
```
