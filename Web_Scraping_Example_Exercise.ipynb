<h3>This Is An Example Question For Web-Scraping An Image From A website And Save It</h3>

```python
import requests
```

<h3>Using Requests Lib To Request The Site We Want To Scrap</h3>

```python
res = requests.get('https://www.holidify.com/places/delhi/waste-to-wonder-park-sightseeing-1254585.html')
```

```python
res.text
```

<h3>Importing Beautiful_Soup Lib To Get The Data From The Website</h3>

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
imag = soup.select('figure > img')[1]
```

```python
img_link = imag['src']
```

<h3>Getting The Image After Finding The Correct Url From The Scraped Data</h3>

```python
img = requests.get('https://www.holidify.com/images/cmsuploads/compressed/52351674_254699908809079_8713762700591198915_n_20190327113023.jpg')
```

<h3>Writing The Contents To A Locally Generated Image File</h3>

```python
with open('Leaning_Tower_Of_Pisa.jpg','wb') as f:
    f.write(img.content)
```

```python
f.close()
```
